```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

#from itertools import islice
```

```python
data = pd.read_excel('../data/telecom_data.xlsx')
```

```python
data_UserEngagement = pd.DataFrame()
data_UserEngagement['msisdn'] = data['MSISDN/Number']
data_UserEngagement['BId'] = data['Bearer Id']
data_UserEngagement['Dur'] = data['Dur. (ms)']
data_UserEngagement['total'] = data['Total DL (Bytes)'] + data['Total UL (Bytes)']

UserEngagement = data_UserEngagement.groupby('msisdn').aggregate({'BId': 'count','Dur': 'sum','total': 'sum',}).rename(columns={"BId": "sessions_freq", "Dur": "sessions_dur(.ms)", "total": "sessions_traff"})
```

```python
UserEngagement
```

# Top 10 users based on session frequency

```python
topUsers_SF = UserEngagement.sort_values(by = 'sessions_freq', ascending = False).head(10)
topUsers_SF['sessions_freq']
```

# Top 10 users based on the total duration(ms) of all the sessions 

```python
topUsers_TD = UserEngagement.sort_values(by = 'sessions_dur(.ms)', ascending = False).head(10)
topUsers_TD['sessions_dur(.ms)']
```

# Top 10 users based on the total traffic(bytes) of all the sessions

```python
topUsers_TS = UserEngagement.sort_values(by = 'sessions_traff', ascending = False).head(10)
topUsers_TS['sessions_traff']
```

```python
normalized_UserEngagement_data= np.log1p(UserEngagement)
normalized_UserEngagement_data.head()
```

```python
Kmean_clustering = KMeans(n_clusters = 3, init = 'k-means++', random_state =0, max_iter = 1000)
Kmean_clustering.fit(normalized_UserEngagement_data)
```
